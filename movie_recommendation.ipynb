```python
import ast
```

```python
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
m['genres']=m['genres'].apply(convert)
m['keywords']=m['keywords'].apply(convert)
#m['cast']=m['cast'].apply(convert)
#m['crew']=m['crew'].apply(convert)
```

```python
def convert3(obj):
  l=[]
  count=0
  for i in ast.literal_eval(obj):
    if count!=3:
      l.append(i['name'])
      count+=1
    else:
      break
  return l
```

```python
m['cast']=m['cast'].apply(convert3)
```

```python
#m['crew'][0]
```

```python
def fetch_dir(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']== 'Director':
      l.append(i['name'])
      break
  return l
```

```python
m['crew']=m['crew'].apply(fetch_dir)
```

```python
m['overview']=m['overview'].apply(lambda x:x.split())
```

```python
m.head(4)
```

```python
m['genres']=m['genres'].apply(lambda x: [i.replace(' ','') for i in x])
m['keywords']=m['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
m['cast']=m['cast'].apply(lambda x: [i.replace(' ','') for i in x])
m['crew']=m['crew'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
m.head(1)
```

```python
m['tags']= m['overview'] + m['genres'] + m['keywords'] + m['cast'] + m['crew']
```

```python
m_new=m[['id','title','tags']]
m_new.head(3)
```

```python
m_new['tags']=m_new['tags'].apply(lambda x:" ".join(x))
```

```python
m_new['tags'][0]
```

```python
m_new['tags']=m_new['tags'].apply(lambda x:x.lower())
```

```python
#for removing similar words like activities, activity and make one similar word
import nltk #natural language toolkit
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
m_new['tags']=m_new['tags'].apply(stem)
```

```python
ps.stem('danced')
```

```python
#making vectors of words in tags
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(m_new['tags']).toarray() #making an array of those vectors
```

```python
vectors[0]
```

```python
#cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
m_new[m_new['title']=='Batman Begins'].index[0]
```

```python
def recommend(movie):
  movie_index=m_new[m_new['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(m_new.iloc[i[0]].title)
    #print(i[0])
  return
```

```python
recommend('Avatar')
```
